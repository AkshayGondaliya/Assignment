```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA 
from sklearn.datasets import load_iris
```

```python
df=load_iris()
```

```python
pca=PCA(n_components=3).fit_transform(df.data)
```

```python
y=df.target
```

```python
ax=Axes3D(plt.figure(1,figsize=(8,8)),elev=-150,azim=110)

ax.scatter(pca[:,0],pca[:,1],pca[:,2],c=y,cmap=plt.cm.rainbow_r)

ax.set_title("PCA")
ax.set_xlabel("1st")
ax.set_ylabel("2nd")
ax.set_zlabel("3rd")

plt.show()
```
